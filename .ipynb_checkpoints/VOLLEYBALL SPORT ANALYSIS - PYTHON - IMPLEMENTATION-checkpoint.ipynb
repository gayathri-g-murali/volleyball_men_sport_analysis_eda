```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('VNL2023.csv')
df.head()
```

```python
# Columns
df.columns
```

```python
# Shape
df.shape
```

```python
# Infomation
df.info()
```

```python
# Summary
df.describe()
```

```python
# Check Null Values
df.isna().sum()
```

```python
# Check Duplicates
df.duplicated().sum()
```

```python
# Check correlation
numeric_cols = df.select_dtypes(include =['int','float']).columns
corr_matrix = df[numeric_cols].corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix,annot = True, cmap = 'cividis', linewidth = 5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Value Counts For Position Columns
position_counts = df['Position'].value_counts()
position_counts
```

```python
plt.figure(figsize=(10,6))
colors = ['#003366','#4682B4','#ADD8E6','#FFD700','#FFFF99']
plt.pie(position_counts,labels = position_counts.index, colors = colors, autopct = '%1.1f%%', startangle = 90)
plt.title('Distribution of Positions')
plt.show()
```

```python
# Group by Average Attack
avg_attack_by_country = df.groupby('Country')['Attack'].mean().sort_values(ascending = False)
avg_attack_by_country
```

```python
plt.figure(figsize=(10,6))
avg_attack_by_country.sort_values(ascending=False).head(10).plot(kind ='bar', color = '#01153E')
plt.xlabel('Country')
plt.ylabel('Average Attack')
plt.title('Average Attack by Countries')
plt.show()
```

```python
# Average Serve by Age
avg_serve_by_age = df.groupby('Age')['Serve'].mean()
avg_serve_by_age.sort_values(ascending=False)
```

```python
# Country by Position and Attack
df.groupby(['Country','Position'])['Attack'].max().reset_index().sort_values(ascending = False, by = 'Attack').head(20)
```

```python
#  Dig by Countries
dig_by_country = df.groupby('Country')['Dig'].sum()
dig_by_country
```

```python
plt.figure(figsize=(10,6))
dig_by_country.sort_values(ascending = False).head(7).plot(kind = 'bar', color = '#01153E')
plt.title('Total dig by Country')
plt.xlabel('Country')
plt.ylabel('Total Dig')
plt.show()
```

```python
# Block v/s Receive
plt.figure(figsize=(10,6))
plt.scatter(df['Block'],df['Receive'], color = '#01153E')
plt.xlabel('Block')
plt.ylabel('Receive')
plt.show()
```

```python
# Boxplot for serve
plt.figure(figsize=(10,6))
sns.boxplot(x = df['Serve'], color = '#01153E')
plt.title('Distribution of Serve values')
plt.xlabel('Serve')
plt.show()
```

```python
# Distribution by age
plt.figure(figsize=(10,6))
plt.hist(df['Age'], bins = 20, color = '#01153E', edgecolor = 'white')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
# Average Attack by Position
avg_attack_by_position = df.groupby('Position')['Attack'].mean()
plt.figure(figsize=(10,6))
avg_attack_by_position.sort_values(ascending=True).plot(kind = 'bar', color = '#01153E')
plt.title('Average Attack by Position')
plt.xlabel('Position')
plt.ylabel('Attack')
plt.show()
```

```python
# Trend of Serve over tha age group
serve_trend_by_age = df.groupby('Age')['Serve'].mean()
plt.figure(figsize=(10,6))
serve_trend_by_age.plot(kind ='line', marker = 'o', linestyle = '-', color = '#01153E')
plt.title('Serve Trend Over Age Group')
plt.xlabel('Age')
plt.ylabel('Serve')
plt.show()
```

```python
# Total attack block by country
total_attack_block_by_country = df.groupby('Country')[['Attack','Block']].sum()
```

```python
total_attack_block_by_country
```

```python
plt.figure(figsize=(10,8))
total_attack_block_by_country.sort_values(ascending=False, by = 'Attack').plot(kind = 'bar', stacked = True, colormap = 'cividis')
plt.title('Total Attack and Block by Country')
plt.xlabel('Country')
plt.ylabel('Total Value')
plt.show()
```
